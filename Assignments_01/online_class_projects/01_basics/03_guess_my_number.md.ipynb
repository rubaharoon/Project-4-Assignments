```python
import random
import time

def guess_my_number():
    print("\n🌟🔮 Welcome, brave challenger! 🔮🌟")
    print("I am the **Mystic Number Keeper**, and I have locked a secret number between **0 and 99** in my vault! 🔐")
    print("Do you have what it takes to guess it? 🧐💡")
    print("Every wrong guess brings you **closer to destiny** or **further into the abyss of mystery**... 🌌")
    print("🎯 Type your best guess and let the game of wits begin! 🎯\n")

    secret_number = random.randint(0, 99)
    attempts = 0

    while True:
        try:
            guess = int(input("🔢 Enter your guess: "))
            attempts += 1

            if guess < 0 or guess > 99:
                print("🚨 Out of range! Please enter a number between 0 and 99.")
            elif guess > secret_number:
                print("📉 Too high! The vault remains locked... Try again! 🔻")
            elif guess < secret_number:
                print("📈 Too low! The secret stays hidden... Try again! 🔺")
            else:
                print("\n✨ **A blinding light flashes! The vault unlocks!** ✨")
                time.sleep(1)
                print(f"🎉 **Congratulations, mighty challenger!** The number was: {secret_number} 🎯")
                print(f"🏆 You cracked the code in **{attempts} attempts**!")
                print("🚀 The Mystic Keeper bows in respect... You have mastered the game! 🎭\n")
                break
        except ValueError:
            print("⚠️ Invalid input! Please enter a number.")

# Run the game
guess_my_number()


```
