```python
import time

def spaceship_launch():
    print("\n🚀 Mission Control: Initiating countdown sequence! 🚀\n")

    for i in range(10, 0, -1):
        print(f"🔢 {i}...", end=" ", flush=True)
        time.sleep(1)

    print("\n🔥💥 3... 2... 1... 💥🔥")
    time.sleep(1)
    print("🚀 LIFTOFF! We have ignition! 🌎✨")

spaceship_launch()
```
