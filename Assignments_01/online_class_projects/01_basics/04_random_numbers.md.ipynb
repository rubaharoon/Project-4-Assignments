```python
import random
import time

print("\n🎤🎉 WELCOME TO THE ULTIMATE NUMBER DRAW! 🎉🎤")
print("🔢 10 lucky numbers will be revealed... Who knows what fate has in store? 🤩\n")

for i in range(1, 11):
    time.sleep(0.7)
    lucky_number = random.randint(1, 100)
    print(f"🎊 Number {i}: 🏆 **{lucky_number}** 🏆")

print("\n🎇✨ That's all, folks! Hope these numbers bring you luck! 🍀🚀\n")
```
