**Bulk File Re-namer Python Project**

```python
import os
from google.colab import files
from datetime import datetime
import mimetypes

# Step 1: Upload files
print("ðŸ“‚âœ¨ Welcome to the File Glow-Up Machine â€” File Type Edition! âœ¨ðŸ“‚")
print("ðŸ”½ Upload your files and watch the categories shine...")
uploaded = files.upload()

# Step 2: Define glow-up styles by file type
glow_styles = {
    'image':   ("ðŸ“¸_ImageGlow", 1),
    'document':("ðŸ“š_DocSparkle", 1),
    'audio':   ("ðŸŽµ_SoundBlast", 1),
    'video':   ("ðŸŽ¬_MovieMagic", 1),
    'other':   ("ðŸŒ€_MysteryFile", 1)
}

def categorize_file(file_name):
    mime, _ = mimetypes.guess_type(file_name)
    if mime:
        if mime.startswith("image"):
            return "image"
        elif mime.startswith("application"):
            return "document"
        elif mime.startswith("audio"):
            return "audio"
        elif mime.startswith("video"):
            return "video"
    return "other"

# Step 3: Rename files with special type-based magic
print("\nðŸ§ª Detecting file types and renaming with purpose...\n")
renamed_files = []

for filename in uploaded.keys():
    file_type = categorize_file(filename)
    base_name, count = glow_styles[file_type]
    name, ext = os.path.splitext(filename)
    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")

    new_name = f"{base_name}_{timestamp}_{count:02d}{ext}"
    glow_styles[file_type] = (base_name, count + 1)

    os.rename(filename, new_name)
    print(f"âœ¨ {filename} âžœ {new_name}")
    renamed_files.append(new_name)

# Step 4: Download renamed glow-up files
print("\nðŸšš Downloading your sparkling new files...\n")
for f in renamed_files:
    files.download(f)

print("ðŸŽ‰ DONE! Each file renamed with its own glow-up style based on type. You're officially organized âœ¨")
```
