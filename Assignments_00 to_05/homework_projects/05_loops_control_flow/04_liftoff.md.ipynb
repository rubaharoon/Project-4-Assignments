```python
import time

def spaceship_launch():
    """Epic countdown for spaceship launch in a pyramid shape!"""
    print("\n🌍🚀 Mission Control: All systems are GO for launch! 🛰️")
    print("🔎 Final checks complete. Standby for countdown...\n")

    countdown = [str(i) for i in range(10, 0, -1)]

    for i in range(1, 11):
        print(" ".join(countdown[:i]))
        time.sleep(1)

    print("\n🔥🚀 LIFTOFF! We have a successful launch! 🌍✨\n")

# Start the countdown sequence
if __name__ == "__main__":
    spaceship_launch()
```
