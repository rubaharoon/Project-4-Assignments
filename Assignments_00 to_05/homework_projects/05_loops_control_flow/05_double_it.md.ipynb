```python
def doubling_challenge():
    """Keeps doubling the number until it reaches or exceeds 100."""
    print("\n🎲 Welcome to The Doubling Challenge! 🚀")
    print("💡 Start with any number, and watch it grow until it reaches 100 or more!\n")

    try:
        curr_value = float(input("🔢 Enter your starting number: "))
    except ValueError:
        print("❌ Invalid input! Please enter a valid number next time.")
        return

    while curr_value < 100:
        print(f"✨ Doubling {curr_value}... 🌀")
        curr_value *= 2

    print(f"\n🎉 Boom! Your number is now {curr_value} — Mission Accomplished! 💥🚀\n")

# Start the game
if __name__ == "__main__":
    doubling_challenge()
```
