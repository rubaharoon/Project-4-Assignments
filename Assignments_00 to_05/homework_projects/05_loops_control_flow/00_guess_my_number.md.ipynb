```python
import random

def mind_reader_game():
    """A fun number guessing game with hints, limited attempts, and an exit option."""
    secret_number = random.randint(0, 99)
    attempts = 7

    print("\n🔮 Welcome to The Mind Reader Challenge! 🎩")
    print("I have chosen a mysterious number between 0 and 99... Can you guess it? 🤔")
    print(f"You have {attempts} attempts. Type 'exit' anytime to quit.\n")

    for attempt in range(1, attempts + 1):
        while True:
            guess = input(f"✨ Attempt {attempt}/{attempts} - Enter your guess: ").strip()

            if guess.lower() == "exit":
                print("🚪 You have chosen to leave the mystical realm... Goodbye, mortal! 🌀👋\n")
                return

            try:
                guess = int(guess)
                if 0 <= guess <= 99:
                    break
                else:
                    print("⚠️ Please guess a number **between 0 and 99!**")
            except ValueError:
                print("❌ Invalid input! Please enter a valid number or type 'exit' to quit.")

        if guess == secret_number:
            print(f"🚀 Correct! 🎉 The number was {secret_number}.")
            print(f"👑 You are a true **Mind Reader!** 🧠🔮\n")
            return
        elif guess > secret_number:
            print("🔥 Too high! Try again.")
        else:
            print("❄️ Too low! Try again.")

    print(f"\n💀 You have exhausted all your attempts! The number was **{secret_number}**.")
    print("🔮 Better luck next time, psychic apprentice! 🕵️‍♂️\n")

# Run the game
if __name__ == "__main__":
    mind_reader_game()
```
