```python
def fibonacci_sequence(max_value=10000):
    """Generates Fibonacci numbers up to the max_value."""
    a, b = 0, 1
    print("\n🐇 Welcome to the Fibonacci Rabbit Race! 🏁")
    print("🌱 The Fibonacci rabbits are multiplying... Here’s the sequence:\n")

    while a < max_value:
        print(f"{a}", end=" ")
        a, b = b, a + b
    print("\n\n🎉 The rabbits have taken over! Fibonacci sequence completed up to 10,000! 🚀🐇")


if __name__ == "__main__":
    fibonacci_sequence()
```
