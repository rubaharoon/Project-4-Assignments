```python
def affirmation_challenge():
    """Encourages the user to type a positive affirmation correctly."""
    affirmation = "I am capable of doing anything I put my mind to."

    print("\n💫 Welcome to The Power of Affirmations! 💪")
    print("🌟 Sometimes we need to remind ourselves how strong we truly are.")
    print("📝 Keep typing this affirmation until you get it right:")
    print(f'👉 "{affirmation}"\n')

    while True:
        user_input = input("> ").strip()

        if user_input == affirmation:
            print("\n🎉 You did it! Never forget your strength! 💖💪")
            print("🚀 Now go conquer your dreams! 🌍✨\n")
            break
        else:
            print("❌ Oops! That's not quite right. Try again! You got this! 🔥")

# Run the affirmation challenge
if __name__ == "__main__":
    affirmation_challenge()
```
