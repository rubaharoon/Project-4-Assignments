```python
def main():

    print("✨ Welcome to the Magic Number Squarer! 🔢✨")
    print("Let's find out the square of your favorite number! 🧐\n")


    number = float(input("🔢 Enter a number: "))

    # Calculate the square
    square = number * number


    print(f"🎉 Wow! The square of {number} is {square}! 🎊")

    print("🧠 Fun Fact: Squaring a number means multiplying it by itself! 🤯 Try a negative number too!")


if __name__ == "__main__":
    main()
```
