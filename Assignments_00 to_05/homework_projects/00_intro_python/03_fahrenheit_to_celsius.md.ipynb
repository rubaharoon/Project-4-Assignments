**03_fahrenheit_to_celsius.md**

```python
def main():

    print("🌡️ Welcome to the Ultimate Temperature Converter! ❄️🔥")
    print("Ever wondered how hot or cold it is in Celsius? Let's find out! 🧐\n")

    fahrenheit = float(input("🌞 Enter the temperature in Fahrenheit: "))

    # Convert Fahrenheit to Celsius
    celsius = (fahrenheit - 32) * 5.0 / 9.0


    print(f"🌍 The temperature {fahrenheit}°F is equal to {celsius:.2f}°C! 🌡️")


    if celsius <= 0:
        print("❄️ Brrr! That's freezing! Better grab a warm coat! 🧥☃️")
    elif celsius >= 30:
        print("🔥 Whoa! That's pretty hot! Stay hydrated and find some shade! 🍹🌴")
    else:
        print("🌤️ Perfect weather! Not too hot, not too cold—just right! 😎")

if __name__ == "__main__":
    main()
```
