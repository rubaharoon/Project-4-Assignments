**01_add_two_numbers.md**

```python
def main():
    print("➕ Welcome to the Magic Adder! 🎩✨")
    print("Let's add two numbers and see the magic happen! 🧮\n")

    num1 = int(input("🔢 Enter the first number: "))

    num2 = int(input("🔢 Enter the second number: "))

    total_sum = num1 + num2

    print(f"🎉 Ta-da! The sum of {num1} and {num2} is {total_sum}! 🎈")
    print("Numbers are amazing, aren't they? 😃")

if __name__ == "__main__":
    main()
```
