```python
def main():
    print("🌍 Welcome to the Wonderful World of Animals! 🦁🐯🐰")
    print("Let's talk about your favorite animal! 🎉\n")

    favorite_animal = input("🧐 What is your favorite animal? ")

    print(f"😃 No way! My favorite animal is also {favorite_animal}! 🎊")

    fun_facts = {
        "dog": "🐶 Did you know? Dogs have an extraordinary sense of smell and can be trained to detect diseases!",
        "cat": "🐱 Fun fact: Cats can make over 100 different sounds! Meow! 🎶",
        "elephant": "🐘 Wow! Elephants have incredible memories and can recognize friends even after decades!",
        "dolphin": "🐬 Amazing! Dolphins sleep with one eye open and half their brain awake! 💤",
        "owl": "🦉 Owls can rotate their heads up to 270 degrees! 😲 Spooky, right?",
    }

    animal_lower = favorite_animal.lower()
    if animal_lower in fun_facts:
        print(fun_facts[animal_lower])
    else:
        print("✨ That's a fantastic choice! Every animal is special in its own way! 🌟")

# Run the main function
if __name__ == "__main__":
    main()
```
