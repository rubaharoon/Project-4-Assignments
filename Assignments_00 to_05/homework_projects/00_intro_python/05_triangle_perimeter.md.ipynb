```python
def main():

    print("📐 Welcome to the Triangle Perimeter Calculator! 🔺✨")
    print("Let's find out the perimeter of your triangle! 🧐\n")


    side1 = float(input("📏 What is the length of side 1? "))
    side2 = float(input("📏 What is the length of side 2? "))
    side3 = float(input("📏 What is the length of side 3? "))

    # Calculate the perimeter
    perimeter = side1 + side2 + side3


    print(f"🎉 The perimeter of the triangle is {perimeter} units! 🎊")


    print("🔺 Fun Fact: The sum of any two sides of a triangle is always greater than the third side! 😉")


if __name__ == "__main__":
    main()
```
