```python
import time


MIN_HEIGHT = 50

print("\n🎢✨ Welcome to the Thrill Seeker Rollercoaster! 🎡🏎️")
print("Before you ride, let's check if you meet the height requirement... ⏳")

while True:

    height = input("\n📏 How tall are you? (Enter a number or press Enter to exit): ")


    if height.strip() == "":
        print("\n👋 Thanks for visiting the theme park! Come back soon! 🎠✨")
        break

    try:
        height = float(height)
        print("\n🔎 Checking height requirements...")
        time.sleep(1)

        if height >= MIN_HEIGHT:
            print(f"\n🎉 Woohoo! At {height} units tall, you're **TALL ENOUGH** to ride! 🎢🔥 Enjoy the thrill! 🏎️💨")
        else:
            print(f"\n😞 Oops! You're {height} units tall, which is **too short** to ride. ⏳")
            print("📈 Come back when you've grown a little taller! We’ll be waiting! 🏰🎠")

    except ValueError:
        print("\n⚠️ Please enter a valid number! Try again. 🔄")

print("\n🎪 The amusement park is closing for now. Have a magical day! ✨🎆")
```
