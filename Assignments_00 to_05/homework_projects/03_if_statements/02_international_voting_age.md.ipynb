```python
import time

# Welcome message
print("🌍✨ Welcome, traveler! You’ve arrived at the Grand Voting Embassy! 🏛️")
time.sleep(1)
age = int(input("\n🎂 How old are you, wise one? "))


countries = {
    "🏛️ Peturksbouipo": 16,
    "🏰 Stanlau": 25,
    "🌿 Mayengua": 48
}

print("\n🔎 Processing your voting eligibility...")
time.sleep(1)


print("\n⏳ Scanning global voting records...\n")
time.sleep(1)


for country, voting_age in countries.items():
    time.sleep(0.8)
    if age >= voting_age:
        print(f"🎉✅ Hooray! You CAN vote in {country} (Minimum age: {voting_age}) 🗳️")
    else:
        print(f"❌ Oh no! You CANNOT vote in {country} yet. You need to be at least {voting_age}. ⏳")


print("\n🌟 Your Voting Passport has been updated! 🌍✈️")
print("💡 Remember: The future belongs to those who participate! 🗳️✨\n")
```
