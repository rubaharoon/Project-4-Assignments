```python
import time


print("\n🌍✨ Welcome to the Leap Year Detector 🚀")
time.sleep(1)


year = int(input("\n📅 Enter a year to check if it's a leap year: "))

print("\n🔎 Analyzing the year...")
time.sleep(1)


if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print(f"\n🎉 Hooray! {year} is a **LEAP YEAR**! 🎊 Enjoy an extra day in February! 🏆")
else:
    print(f"\n❌ Nope! {year} is **NOT a leap year**. Better luck next time! ⏳")


print("\n💡 Fun Fact: Leap years were introduced by Julius Caesar in 45 BCE to keep the calendar aligned with Earth's revolutions! 🌍📜")
```
