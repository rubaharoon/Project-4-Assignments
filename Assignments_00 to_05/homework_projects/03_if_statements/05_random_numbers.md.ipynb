```python
import random
import time


print("\n🎰 Welcome to the Lucky 10 Random Draw! 🎲✨")
time.sleep(1)
print("🔢 Let's generate **10 random numbers** between **1 and 100**...\n")
time.sleep(1)


random_numbers = []
for i in range(10):
    num = random.randint(1, 100)
    random_numbers.append(num)


    print(f"🎉 Lucky Number {i+1}: {num} 🎯")
    time.sleep(0.8)


print("\n🎆✨ Here are your **lucky 10 numbers**: ✨🎆")
print("🎲", " - ".join(map(str, random_numbers)), "🎲")


print("\n💡 Fun Fact: Did you know? The number 7 is considered lucky in many cultures! 🍀😃")
```
