**01_add_many_number.md**

```python
import time

def sum_numbers(numbers, index=0, total=0):
    if index == len(numbers):
        print("\n✨ Drumroll... The final sum is:", total, "✨")
        return total
    else:
        print(f"Adding {numbers[index]}... Current total: {total + numbers[index]}")
        time.sleep(0.5)
        return sum_numbers(numbers, index + 1, total + numbers[index])


nums = [3, 7, 6, 9, 5]
sum_numbers(nums)
```
