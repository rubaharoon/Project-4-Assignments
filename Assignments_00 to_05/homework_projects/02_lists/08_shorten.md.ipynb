```python
import time

MAX_LENGTH = 3

def typing_effect(text, delay=0.05):
    """Simulates a typing effect for added fun!"""
    for char in text:
        print(char, end="", flush=True)
        time.sleep(delay)
    print()

def shorten(lst):
    """Removes elements from lst until it is MAX_LENGTH long."""
    typing_effect("\nğŸ› ï¸ Checking the list size...")

    if len(lst) <= MAX_LENGTH:
        typing_effect("âœ… The list is already within the limit! No changes needed. ğŸ‰")
        return

    typing_effect(f"âš ï¸ List is too long! Trimming down to {MAX_LENGTH} items...\n")

    while len(lst) > MAX_LENGTH:
        removed_item = lst.pop()
        typing_effect(f"ğŸ”¥ Shredding '{removed_item}'... âœ‚ï¸")
        time.sleep(0.5)

    typing_effect("\nğŸ Done! Your shortened list is:")
    print(f"ğŸ“œ {lst} âœ…\n")

def main():
    """Gets a list from the user and passes it to shorten()."""
    user_list = []

    typing_effect("\nğŸ“œ Welcome to the List Trimmer!")
    typing_effect("âœ¨ Keep entering values to build your list.")
    typing_effect(f"ğŸ”¹ The list can only have {MAX_LENGTH} items max!")
    typing_effect("ğŸ”¹ Press Enter without typing anything to finish.\n")

    while True:
        item = input("â¡ï¸ Enter a value: ")
        if item == "":
            break
        user_list.append(item)
        typing_effect(f"âœ… '{item}' added to the list!")

    typing_effect("\nğŸ” Analyzing your list...\n", delay=0.1)
    time.sleep(1)

    shorten(user_list)
    typing_effect("\nğŸ‰ Thank you for using the List Trimmer! Goodbye!\n")

if __name__ == "__main__":
   main()
```
