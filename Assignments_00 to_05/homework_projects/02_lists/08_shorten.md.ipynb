```python
import time

MAX_LENGTH = 3

def typing_effect(text, delay=0.05):
    """Simulates a typing effect for added fun!"""
    for char in text:
        print(char, end="", flush=True)
        time.sleep(delay)
    print()

def shorten(lst):
    """Removes elements from lst until it is MAX_LENGTH long."""
    typing_effect("\n🛠️ Checking the list size...")

    if len(lst) <= MAX_LENGTH:
        typing_effect("✅ The list is already within the limit! No changes needed. 🎉")
        return

    typing_effect(f"⚠️ List is too long! Trimming down to {MAX_LENGTH} items...\n")

    while len(lst) > MAX_LENGTH:
        removed_item = lst.pop()
        typing_effect(f"🔥 Shredding '{removed_item}'... ✂️")
        time.sleep(0.5)

    typing_effect("\n🎁 Done! Your shortened list is:")
    print(f"📜 {lst} ✅\n")

def main():
    """Gets a list from the user and passes it to shorten()."""
    user_list = []

    typing_effect("\n📜 Welcome to the List Trimmer!")
    typing_effect("✨ Keep entering values to build your list.")
    typing_effect(f"🔹 The list can only have {MAX_LENGTH} items max!")
    typing_effect("🔹 Press Enter without typing anything to finish.\n")

    while True:
        item = input("➡️ Enter a value: ")
        if item == "":
            break
        user_list.append(item)
        typing_effect(f"✅ '{item}' added to the list!")

    typing_effect("\n🔍 Analyzing your list...\n", delay=0.1)
    time.sleep(1)

    shorten(user_list)
    typing_effect("\n🎉 Thank you for using the List Trimmer! Goodbye!\n")

if __name__ == "__main__":
   main()
```
