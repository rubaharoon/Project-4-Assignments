```python
import time

def get_last_element(lst):
    """Prints the last element of a non-empty list."""
    print("\n🔮 Activating time machine to find the last element...")
    time.sleep(1)

    print("\n🚀 Traveling through the list dimensions...")
    for _ in range(3):
        print("⏳ Whoosh! Zooming forward...")
        time.sleep(0.5)

    print(f"\n✨ Destination reached! The last element is: **{lst[-1]}** 🎉\n")


user_list = []
n = int(input("🔢 How many elements do you want to enter? "))

print("\n📝 Enter your elements one by one:")

for i in range(n):
    item = input(f"➡️ Element {i + 1}: ")
    user_list.append(item)


get_last_element(user_list)
```
