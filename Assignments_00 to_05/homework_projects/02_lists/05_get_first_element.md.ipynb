```python
import time

def get_first_element(lst):
    """Prints the first element of a non-empty list."""
    print("\n🔍 Searching for the first element...")
    time.sleep(1)

    print("\n🎁 Unwrapping the mystery box...")
    for _ in range(3):
        print("🕵️‍♂️ Looking inside...")
        time.sleep(0.5)

    print(f"\n✨ Ta-da! The first element is: **{lst[0]}** 🎉\n")


user_list = []
n = int(input("🔢 How many elements do you want to enter? "))

print("\n📝 Enter your elements one by one:")

for i in range(n):
    item = input(f"➡️ Element {i + 1}: ")
    user_list.append(item)


get_first_element(user_list)
```
