```python
import time

def add_three_copies(my_list, item):
    """Adds three copies of item to the given list (mutable)."""
    print(f"\n📌 Adding '{item}' three times to the list...\n")
    time.sleep(0.5)

    for i in range(3):
        my_list.append(item)
        print(f"✅ Added copy {i + 1}...")
        time.sleep(0.5)

    print("\n🎉 Done! Notice how we didn’t return anything, but the list changed!\n")

def change_number(x):
    """Attempts to change an immutable type (number)."""
    print(f"\n🔄 Changing {x} to {x + 10} inside the function...")
    time.sleep(0.5)
    x += 10
    print("❌ The change is lost once we leave the function!\n")


items_list = []


user_input = input("\n🔢 Enter a number or word to add to the list: ")
add_three_copies(items_list, user_input)

print(f"📜 Final list: {items_list} (Notice the change stayed!)\n")

x = 5
print(f"\n🧮 Original number: {x}")
change_number(x)
print(f"📌 After function call: {x} (No change!)\n")

print("🔍 **Mutable types (lists) keep changes, but immutable types (numbers) do not!** 🚀")
```
