```python
import time

def typing_effect(text, delay=0.05):
    """Simulates a typing effect for a given text."""
    for char in text:
        print(char, end="", flush=True)
        time.sleep(delay)
    print()

def collect_inputs():
    """Continuously asks for input until the user presses Enter (empty input)."""
    user_list = []

    typing_effect("\n🕵️ Welcome to the Secret Code Compiler! 🔐")
    typing_effect("🔎 Your mission: Securely compile classified data.")
    typing_effect("💻 Keep entering intel—each entry is encrypted in the mainframe.")
    typing_effect("🛑 Press Enter without typing anything to finalize the database.\n")

    while True:
        item = input("🔏 Enter classified data: ")
        if item == "":
            break
        user_list.append(item)
        typing_effect(f"✅ '{item}' has been encoded into the system!")

    typing_effect("\n🔄 Running final encryption protocol...\n", delay=0.1)
    time.sleep(1)

    typing_effect("🔐 Data successfully locked in the classified vault! 🛡️")
    print(f"\n📁 Your Secret Archive: {user_list}\n")

# Run the program
collect_inputs()
```
