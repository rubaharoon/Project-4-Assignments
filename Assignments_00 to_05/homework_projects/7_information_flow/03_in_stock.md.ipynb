```python
import time
import random

# Sophia's inventory 📦
inventory = {
    "apple": 10,
    "banana": 5,
    "orange": 8,
    "grape": 2,
    "mango": 7,
    "strawberry": 0,
    "pineapple": 3
}

def num_in_stock(fruit):
    return inventory.get(fruit.lower(), 0)

def main():
    print("\n🍎 Welcome to **Sophia's Amazing Fruit Store!** 🍌")
    time.sleep(1)

    suspense = [
        "🔍 Checking the shelves...",
        "📦 Searching through the inventory...",
        "⏳ Almost there..."
    ]

    for s in suspense:
        print(s)
        time.sleep(random.uniform(0.5, 1.2))

    fruit = input("\n🛒 Enter a fruit: ").lower()
    stock = num_in_stock(fruit)

    if stock > 0:
        print(f"\n✅ Great news! We have **{stock} {fruit}(s)** in stock! 🥳")
    else:
        print(f"\n❌ Oh no! **{fruit}** is currently **out of stock**. 😞 Try another fruit!")

# Run the program
main()
```
