```python
import time
import random

# Sophia's inventory ğŸ“¦
inventory = {
    "apple": 10,
    "banana": 5,
    "orange": 8,
    "grape": 2,
    "mango": 7,
    "strawberry": 0,
    "pineapple": 3
}

def num_in_stock(fruit):
    return inventory.get(fruit.lower(), 0)

def main():
    print("\nğŸ Welcome to **Sophia's Amazing Fruit Store!** ğŸŒ")
    time.sleep(1)

    suspense = [
        "ğŸ” Checking the shelves...",
        "ğŸ“¦ Searching through the inventory...",
        "â³ Almost there..."
    ]

    for s in suspense:
        print(s)
        time.sleep(random.uniform(0.5, 1.2))

    fruit = input("\nğŸ›’ Enter a fruit: ").lower()
    stock = num_in_stock(fruit)

    if stock > 0:
        print(f"\nâœ… Great news! We have **{stock} {fruit}(s)** in stock! ğŸ¥³")
    else:
        print(f"\nâŒ Oh no! **{fruit}** is currently **out of stock**. ğŸ˜ Try another fruit!")

# Run the program
main()
```
