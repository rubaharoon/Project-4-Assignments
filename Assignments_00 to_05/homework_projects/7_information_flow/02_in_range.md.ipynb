```python
import time
import random

def in_range(n, low, high):
    print("\n🔎 Scanning the number through the range analyzer... 🤖")
    time.sleep(1)

    suspense = [
        "📊 Evaluating boundaries...",
        "🧐 Checking if the number fits within the limits...",
        "⏳ Calculating results..."
    ]

    for s in suspense:
        print(s)
        time.sleep(random.uniform(0.5, 1.2))

    if low <= n <= high:
        print(f"\n✅ Success! {n} is **within** the range of {low} to {high}. 🎯")
        return True
    else:
        print(f"\n❌ Alert! {n} is **outside** the range of {low} to {high}. 🚧")
        return False

def main():
    print("\n🎯 Welcome to the Range Checker 3000! 🚀")
    time.sleep(1)

    n = int(input("\n🔢 Enter a number to check: "))
    low = int(input("⬇ Enter the lower bound: "))
    high = int(input("⬆ Enter the upper bound: "))

    in_range(n, low, high)

# Run the program
main()
```
