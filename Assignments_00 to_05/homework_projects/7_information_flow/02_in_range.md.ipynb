```python
import time
import random

def in_range(n, low, high):
    print("\nğŸ” Scanning the number through the range analyzer... ğŸ¤–")
    time.sleep(1)

    suspense = [
        "ğŸ“Š Evaluating boundaries...",
        "ğŸ§ Checking if the number fits within the limits...",
        "â³ Calculating results..."
    ]

    for s in suspense:
        print(s)
        time.sleep(random.uniform(0.5, 1.2))

    if low <= n <= high:
        print(f"\nâœ… Success! {n} is **within** the range of {low} to {high}. ğŸ¯")
        return True
    else:
        print(f"\nâŒ Alert! {n} is **outside** the range of {low} to {high}. ğŸš§")
        return False

def main():
    print("\nğŸ¯ Welcome to the Range Checker 3000! ğŸš€")
    time.sleep(1)

    n = int(input("\nğŸ”¢ Enter a number to check: "))
    low = int(input("â¬‡ Enter the lower bound: "))
    high = int(input("â¬† Enter the upper bound: "))

    in_range(n, low, high)

# Run the program
main()
```
