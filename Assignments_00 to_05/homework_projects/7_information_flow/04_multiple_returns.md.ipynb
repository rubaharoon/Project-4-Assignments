```python
import time
import random

def get_user_data():
    print("\n🌐 Welcome to the **FutureNet Signup Portal!** 🚀")
    time.sleep(1)

    suspense = [
        "🔍 Initializing secure connection...",
        "📜 Loading user registration form...",
        "🛡️ Ensuring encrypted data transfer..."
    ]

    for s in suspense:
        print(s)
        time.sleep(random.uniform(0.5, 1.2))


    first_name = input("\n✍️ Enter your **First Name**: ")
    last_name = input("✍️ Enter your **Last Name**: ")
    email = input("📧 Enter your **Email Address**: ")

    print("\n✅ Processing your credentials... Please wait... ⏳")
    time.sleep(1)

    print(f"\n🎉 Welcome, **{first_name} {last_name}**! Your account with **{email}** has been created successfully! 🚀")

    return first_name, last_name, email

def main():
    user_data = get_user_data()
    print("\n📂 User Data Stored Securely: ", user_data)

# Run the program
main()
```
