```python
import time
import random

ADULT_AGE = 18

def is_adult(age):
    print("\n🔎 Analyzing your age for adulthood status... 🤔")
    time.sleep(1)

    suspense = [
        "💭 Checking birth records...",
        "🗂️ Verifying life experience...",
        "📊 Running maturity calculations..."
    ]
    for s in suspense:
        print(s)
        time.sleep(random.uniform(0.5, 1.2))

    if age >= ADULT_AGE:
        print(f"\n🎉 Congratulations! At **{age} years old**, you are officially an ADULT! 🚀💼")
        return True
    else:
        print(f"\n🍼 Oh no! At **{age} years old**, you're still a YOUNGLING! 👶 Enjoy your youth! 🎈")
        return False

def main():
    age = int(input("\n🔢 Enter your age: "))
    is_adult(age)

# Run the program
main()
```
