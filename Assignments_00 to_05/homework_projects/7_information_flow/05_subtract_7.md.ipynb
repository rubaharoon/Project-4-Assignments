```python
import time
import random

def subtract_seven(num):
    print("\n🔢 Initiating the **Mysterious Subtraction Protocol**... ⏳")
    time.sleep(1)

    suspense = [
        "🧠 Calculating the impact of subtracting 7...",
        "🔍 Running advanced subtraction algorithms...",
        "⚡ Activating number reduction sequence..."
    ]

    for s in suspense:
        print(s)
        time.sleep(random.uniform(0.5, 1.2))

    result = num - 7
    print(f"\n🎉 Boom! {num} - 7 = **{result}**! Magic, right? ✨")


    reactions = [
        "🧐 Who knew math could be so thrilling?!",
        "🎭 That was **dramatic**, wasn't it?",
        "🤔 Bet you didn’t expect such an epic subtraction!"
    ]

    print(random.choice(reactions))
    return result

def main():
    print("\n🎩 Welcome to the **Ultimate Subtraction Showdown!** 🔥")
    time.sleep(1)

    num = int(input("\n🔢 Enter a number to subtract 7 from: "))
    subtract_seven(num)

# Run the program
main()
```
