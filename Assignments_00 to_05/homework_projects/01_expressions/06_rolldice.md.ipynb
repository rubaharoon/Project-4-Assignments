```python
import random
import time

def roll_dice():
    """Rolls two dice and returns their values and the total."""
    die1 = random.randint(1, 6)
    die2 = random.randint(1, 6)
    total = die1 + die2
    return die1, die2, total

def get_reaction(total):
    """Returns a fun reaction based on the total roll."""
    reactions = {
        2: "🐍 Oof! Snake Eyes! Tough luck!",
        7: "🍀 Lucky 7! Feeling lucky?",
        11: "🎉 Yo! Eleven! This one's special!",
        12: "🔥 Double sixes! Jackpot roll!",
        3: "😬 Ouch, a three... not your day?",
        4: "👍 A four! Not bad, keep rolling!"
    }
    return reactions.get(total, "🎲 Keep rolling! Anything can happen!")

def main():
    print("🎲 Welcome to the Dice Rolling Arena! 🎲")
    print("Rolling the dice... 🎲🎲\n")

    time.sleep(1)

    # Roll the dice
    die1, die2, total = roll_dice()

    # Print results
    print(f"🎲 First die: {die1}")
    print(f"🎲 Second die: {die2}")
    print(f"✨ Total roll: {total} ✨\n")

    # Print reaction
    print(get_reaction(total))

if __name__ == "__main__":
    main()
```
