```python
# Constant value for the speed of light in meters per second
C = 299792458

def calculate_energy(mass):
    """Calculates energy using Einstein's equation E = m * c^2"""
    return mass * C**2

def main():
    print("🌟 Welcome to Einstein's Energy Calculator! 🌟")
    print("Enter mass in kilograms to calculate energy.")
    print("Type 'exit' to quit.\n")

    while True:
        user_input = input("Enter mass (kg): ").strip()

        if user_input.lower() == "exit":
            print("👋 Exiting program. Have an energetic day!")
            break

        try:
            mass = float(user_input)
            if mass < 0:
                print("❌ Mass cannot be negative. Try again.")
                continue

            energy = calculate_energy(mass)
            print(f"⚡ Equivalent Energy: {energy:.2e} Joules\n")

        except ValueError:
            print("❌ Invalid input! Please enter a numeric mass.\n")

# Run the program
main()
```
