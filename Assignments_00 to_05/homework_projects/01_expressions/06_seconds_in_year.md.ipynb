**06_seconds_in_year.md**

```python

DAYS_IN_YEAR = 365
HOURS_IN_DAY = 24
MINUTES_IN_HOUR = 60
SECONDS_IN_MINUTE = 60

# Calculate seconds in a year
seconds_in_year = DAYS_IN_YEAR * HOURS_IN_DAY * MINUTES_IN_HOUR * SECONDS_IN_MINUTE

# Print the result
print("\n⏳ Ever wondered how many seconds are in a year? Let's do the math!\n")
print("📅 Days in a year:      ", DAYS_IN_YEAR)
print("⏰ Hours in a day:      ", HOURS_IN_DAY)
print("⌛ Minutes in an hour:  ", MINUTES_IN_HOUR)
print("⏳ Seconds in a minute: ", SECONDS_IN_MINUTE)
print("\n🔢 Crunching the numbers... 🧮\n")

# Display the final result
print(f"✨ There are {seconds_in_year:,} seconds in a year! ✨\n")
print("🎉 That's a lot of time! Make every second count! ⏳🚀")
```
