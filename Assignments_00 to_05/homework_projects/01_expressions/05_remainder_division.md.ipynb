```python
def main():
    print("🤖 Welcome to the Magical Divider! ✨")
    print("Give me two numbers, and I'll reveal the division results...")

    while True:
        try:

            num1 = float(input("\n🔢 Enter the first number: "))
            num2 = float(input("🔢 Enter the second number: "))


            if num2 == 0:
                print("🚨 Oops! Division by zero is not allowed. Try again.")
                continue

            quotient = num1 / num2
            remainder = num1 % num2


            print("\n✨ Calculating the magic... 🪄💫\n")
            print(f"📊 {num1} ÷ {num2} = {quotient:.2f}")
            print(f"🧩 The remainder is: {remainder:.2f}")


            if remainder == 0:
                print("🎉 Wow! It's a perfect division. No remainder at all!")
            elif remainder > 0.5 * num2:
                print("🤔 That's a pretty big remainder! Almost half the divisor.")
            else:
                print("🔢 The remainder is small. Almost a perfect division!")

            break

        except ValueError:
            print("❌ Invalid input! Please enter numeric values.")

# Run the program
main()
```
