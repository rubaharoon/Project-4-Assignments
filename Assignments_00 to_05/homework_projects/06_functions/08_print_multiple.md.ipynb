```python
import time

def print_multiple(message, repeats):
    print("\n🔊 Get ready for the ultimate message spam! 🚀\n")
    time.sleep(1)

    if repeats <= 0:
        print("❌ Oops! No repeats? That's no fun! Try again with a positive number.")
        return

    for i in range(1, repeats + 1):
        print(f"📢 ({i}/{repeats}) {message}")
        time.sleep(0.5)

    print("\n🎉 Boom! Message spam complete! Hope you enjoyed that! 😆\n")

def main():
    message = input("\n💬 Enter a message to repeat: ")
    repeats = int(input("🔄 Enter the number of times to repeat: "))
    print_multiple(message, repeats)

# Run the program
main()
```
