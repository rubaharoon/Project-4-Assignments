```python
import time
import random

def make_sentence(word, part_of_speech):
    print("\n✨ Generating an epic sentence... ✨")
    time.sleep(1)

    templates = {
        0: f"📦 I am excited to add this **{word}** to my vast collection of them! 🏆",
        1: f"🏃 It's so nice outside today, it makes me want to **{word}**! 🌞",
        2: f"🌅 Looking out my window, the sky is big and **{word}**! 🌈"
    }

    if part_of_speech not in templates:
        print("❌ Oops! Invalid part of speech. Try again with 0 (noun), 1 (verb), or 2 (adjective).")
        return

    suspense = ["💭 Thinking...", "📝 Crafting the perfect sentence...", "🎭 Almost ready!"]
    for s in suspense:
        print(s)
        time.sleep(random.uniform(0.5, 1.2))

    print(f"\n{templates[part_of_speech]}\n🎉 Hope you like your new sentence!\n")

def main():
    word = input("\n🔤 Enter a word: ")
    part_of_speech = int(input("🔢 Enter 0 for noun, 1 for verb, or 2 for adjective: "))
    make_sentence(word, part_of_speech)

# Run the program
main()
```
