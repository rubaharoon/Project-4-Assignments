```python
import random
import time

DONE_LIKELIHOOD = 0.3

def done():
    return random.random() < DONE_LIKELIHOOD

def chaotic_counting():
    print("\nâš¡ Initiating Chaotic Counting! âš¡")
    time.sleep(1)

    for num in range(1, 11):
        if done():
            print(f"\nðŸš¨ Chaos strikes at {num}! Stopping abruptly! ðŸš¨")
            return

        print(f"ðŸ”¢ Counting: {num}")
        time.sleep(random.uniform(0.3, 1.2))

    print("\nðŸŽ‰ We made it to 10! Against all odds! ðŸŽ‰")

def main():
    chaotic_counting()
    print("\nI'm done.")

# Run the program
main()
```
