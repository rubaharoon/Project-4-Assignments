```python
import time
import random

def double(num):
    print("\n⚡ Preparing to unleash the power of doubling! ⚡")
    time.sleep(1)

    suspense = ["🔄 Calculating...", "✨ Enhancing...", "🚀 Multiplying..."]
    for s in suspense:
        print(s)
        time.sleep(random.uniform(0.5, 1.5))

    result = num * 2


    if result < 10:
        reaction = "🐣 A tiny double! But still mighty!"
    elif result < 100:
        reaction = "🔥 That's a solid double! Keep going!"
    else:
        reaction = "🚀 Whoa! We've gone into hyperspace with that number!"

    print(f"\n💥 Boom! {num} doubled is {result}! {reaction}\n")
    return result

def main():
    user_input = input("\nEnter a number to double: ")

    try:
        num = float(user_input)
        double(num)
    except ValueError:
        print("\n❌ That's not a valid number! Try again with a real number.")

# Run the program
main()
```
