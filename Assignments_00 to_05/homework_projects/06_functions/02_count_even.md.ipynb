```python
import time
import random

def count_even():
    print("\nğŸ² Welcome to the Even Number Counter! ğŸ²\n")

    numbers = []

    while True:
        user_input = input("Enter an integer or press enter to stop: ")

        if user_input == "":
            break

        try:
            number = int(user_input)
            numbers.append(number)
            print(f"âœ… {number} added! Keep going!")
        except ValueError:
            print("âŒ Oops! That's not a valid integer. Try again!")


    even_count = sum(1 for num in numbers if num % 2 == 0)


    time.sleep(1)
    print("\nğŸ” Analyzing your numbers...")
    time.sleep(random.uniform(0.5, 1.5))

    if even_count == 0:
        print("\nğŸ˜± Oh no! No even numbers found. Pure chaos!")
    elif even_count < len(numbers) // 2:
        print(f"\nğŸ¤” Hmm... Only {even_count} evens? The odds are winning!")
    else:
        print(f"\nğŸ‰ Whoa! You found {even_count} even numbers! Evenness prevails!")

# Run the function
count_even()
```
