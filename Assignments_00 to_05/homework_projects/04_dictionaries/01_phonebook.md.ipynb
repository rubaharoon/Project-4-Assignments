```python
def display_menu():
    """Displays the main menu options."""
    print("\nüïµÔ∏è Secret Agent Phonebook: Mission Control")
    print("=" * 40)
    print("1Ô∏è‚É£ Add a new agent")
    print("2Ô∏è‚É£ Search for an agent")
    print("3Ô∏è‚É£ View all agents")
    print("4Ô∏è‚É£ Remove an agent")
    print("5Ô∏è‚É£ Exit")

def add_agent(phonebook):
    """Adds a new agent to the phonebook."""
    name = input("Enter agent's name: ").strip()
    number = input("Enter agent's number: ").strip()

    if name in phonebook:
        print(f"‚ö†Ô∏è Agent {name} is already in the phonebook!")
    else:
        phonebook[name] = number
        print(f"‚úÖ Agent {name} has been added!")

def search_agent(phonebook):
    """Searches for an agent in the phonebook."""
    name = input("Enter agent's name: ").strip()

    if name in phonebook:
        print(f"üìû {name}'s number: {phonebook[name]}")
    else:
        print(f"‚ùå Agent {name} not found in the system!")

def view_agents(phonebook):
    """Displays all agents in the phonebook."""
    if phonebook:
        print("\nüìú Current Agents:")
        for name, number in phonebook.items():
            print(f" - {name}: {number}")
    else:
        print("üì≠ No agents found. Your list is empty!")

def remove_agent(phonebook):
    """Removes an agent from the phonebook."""
    name = input("Enter agent's name to remove: ").strip()

    if name in phonebook:
        del phonebook[name]
        print(f"üö® Agent {name} has been erased from records!")
    else:
        print(f"‚ùå No record found for Agent {name}!")

def main():
    """Runs the secret agent phonebook system."""
    phonebook = {}

    while True:
        display_menu()
        choice = input("\nEnter your choice: ").strip()

        if choice == "1":
            add_agent(phonebook)
        elif choice == "2":
            search_agent(phonebook)
        elif choice == "3":
            view_agents(phonebook)
        elif choice == "4":
            remove_agent(phonebook)
        elif choice == "5":
            print("üîí Mission Complete. Exiting secure mode.")
            break
        else:
            print("‚ùó Invalid choice! Please enter a number between 1-5.")

# Run the program
if __name__ == "__main__":
    main()
```
