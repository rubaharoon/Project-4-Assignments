```python
import hashlib

def hash_password(password):
    """Converts a password into a secure hash."""
    return hashlib.sha256(password.encode()).hexdigest()

def register_user(stored_logins):
    """Registers a new user with a hashed password."""
    email = input("📩 Enter email: ").strip()
    password = input("🔑 Enter password: ").strip()

    if email in stored_logins:
        print("⚠️ Email already registered! Try logging in.")
    else:
        stored_logins[email] = hash_password(password)
        print("✅ Registration successful! Your password is securely stored.")

def login(stored_logins):
    """Checks if the entered email and password match the stored hash."""
    email = input("\n📩 Enter email: ").strip()
    password_to_check = input("🔑 Enter password: ").strip()

    if email in stored_logins:
        if stored_logins[email] == hash_password(password_to_check):
            print(f"✅ Access granted! Welcome, Agent {email.split('@')[0]}. 🕵️‍♂️")
        else:
            print("🚨 Access Denied! Intruder alert! 🔥")
    else:
        print("❌ No account found! Please register first.")

def main():
    """Main function to handle user registration and login."""
    stored_logins = {}

    print("🔐 Welcome to CyberVault: Secure Login System 🔐")

    while True:
        print("\n🔹 Choose an option:")
        print("1️⃣ Register New User")
        print("2️⃣ Login")
        print("3️⃣ Exit")

        choice = input("Enter your choice: ").strip()

        if choice == "1":
            register_user(stored_logins)
        elif choice == "2":
            login(stored_logins)
        elif choice == "3":
            print("🔒 CyberVault shutting down. Stay secure! 🛡️")
            break
        else:
            print("⚠️ Invalid choice! Please select 1, 2, or 3.")

# Run the program
if __name__ == "__main__":
    main()
```
