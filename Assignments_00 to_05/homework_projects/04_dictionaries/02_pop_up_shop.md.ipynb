```python
def display_fruit_prices(fruit_prices):
    """Displays the available fruits and their prices."""
    print("\n📜 Fruit Price List:")
    for fruit, price in fruit_prices.items():
        print(f"  🍉 {fruit} - ${price:.2f} each")

def get_fruit_quantities(fruit_prices):
    """Asks the user how many of each fruit they want to buy."""
    quantities = {}
    print("\n🛒 Time to shop! Tell me how many of each fruit you’d like.")

    for fruit in fruit_prices:
        while True:
            try:
                qty = int(input(f"🍉 How many {fruit}s? ").strip())
                if qty < 0:
                    print("❌ Please enter a positive number!")
                else:
                    quantities[fruit] = qty
                    break
            except ValueError:
                print("⚠️ Invalid input! Please enter a valid number.")

    return quantities

def calculate_total_cost(fruit_prices, quantities):
    """Calculates the total cost of all selected fruits."""
    total_cost = 0
    print("\n🛍️ **Shopping Cart:**")

    for fruit, qty in quantities.items():
        if qty > 0:
            cost = fruit_prices[fruit] * qty
            total_cost += cost
            print(f"  {qty} x 🍉 {fruit}s = ${cost:.2f}")

    return total_cost

def main():
    """Main function to run the fruit shop program."""
    print("🍉 Welcome to Berry’s Fruit Market! 🍍")
    print("-" * 40)

    fruit_prices = {
        "Apple": 2.00,
        "Banana": 1.00,
        "Orange": 1.50,
        "Grapes": 3.00,
        "Watermelon": 5.00
    }

    display_fruit_prices(fruit_prices)
    quantities = get_fruit_quantities(fruit_prices)
    total_cost = calculate_total_cost(fruit_prices, quantities)

    print(f"\n💰 **Total Cost: ${total_cost:.2f}**")
    print("\n🎉 Thank you for shopping at Berry’s Fruit Market! Have a fruity day! 🍓")

# Run the program
if __name__ == "__main__":
    main()
```
