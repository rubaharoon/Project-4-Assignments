```python
def get_numbers():
    """Gets user input and converts it into a list of integers."""
    numbers = input("ğŸ‘® Enter the list of numbers (separated by spaces): ").split()
    return [int(num) for num in numbers]

def count_occurrences(numbers):
    """Counts occurrences of each number in the list using a dictionary."""
    number_counts = {}
    for num in numbers:
        number_counts[num] = number_counts.get(num, 0) + 1
    return number_counts

def display_results(number_counts):
    """Displays the count of each number in a fun detective-style format."""
    print("\nğŸ“œ Investigation Report:")
    for num, count in number_counts.items():
        print(f"  - **Suspect {num}** â ğŸ” Seen **{count}** times")

def main():
    """Main function that runs the Number Detective Agency program."""
    print("ğŸ” Welcome to the Number Detective Agency!")
    print("ğŸš” We have a case of repeating suspects. Let's count them!\n")

    numbers = get_numbers()
    number_counts = count_occurrences(numbers)
    display_results(number_counts)

    print("\nğŸ¯ Case Solved! The numbers have been caught and counted. Want to investigate another case? ğŸ•µï¸â€â™‚ï¸")

# Run the program
if __name__ == "__main__":
    main()
```
